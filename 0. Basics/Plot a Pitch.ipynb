```python
import matplotlib.pyplot as plt
from mplsoccer import Pitch, VerticalPitch
```

## Draw a pitch on a new axis

```python
# create a pitch object
pitch = Pitch()
# specify figure size (width, height)
fig, ax = pitch.draw(figsize=(8, 4))
```

## Draw on an existing axis

draw a pitch alongside another matplotlib figure

```python
fig, axs = plt.subplots(nrows=1, ncols=2)
pitch = Pitch()
pie = axs[0].pie(x=[1,2])
pitch.draw(ax=axs[1])
```

## Supported data providers

mplsoccer supports 10 pitch types by specifying the `pitch_type` argument: ‘statsbomb’, ‘opta’, ‘tracab’, ‘wyscout’, ‘uefa’, ‘metricasports’, ‘custom’, ‘skillcorner’, ‘secondspectrum’ and ‘impect’. If you are using tracking data or the custom pitch (‘metricasports’, ‘tracab’, ‘skillcorner’, ‘secondspectrum’ or ‘custom’), you also need to specify the `pitch_length` and `pitch_width`, which are typically 105 and 68 respectively.

```python
pitch = Pitch(pitch_type='wyscout')
fig, ax=pitch.draw()
```

```python
pitch = Pitch(pitch_type='metricasports',  # example plotting a tracab pitch
              pitch_length=105, pitch_width=68,
            axis=True,label=True)  # showing axis labels is optional
fig, ax = pitch.draw()
```

## Adjusting the plot layout

the default is to use tight_layout

```python
pitch = Pitch('statsbomb')
fig, axs = pitch.draw(nrows=2, ncols=2)
```

use constrained_layout

```python
pitch = Pitch()
fig, axs = pitch.draw(nrows=2, ncols=2, tight_layout=False, constrained_layout=True)
```

## Pitch orientation

### 1. Horizontal full

```python
pitch = Pitch()
fig,ax=pitch.draw()
```

### 2. Vertical full

```python
pitch = VerticalPitch()
fig,ax = pitch.draw()
```

### 3. Horizontal half

```python
pitch = Pitch(half=True)
fig, ax = pitch.draw()
```

### 4. Vertical half

```python
pitch = VerticalPitch(half=True)
fig,ax = pitch.draw()
```

adjust the pitch orientations with the pad_left, pad_right, pad_bottom and pad_top arguments to make arbitrary pitch shapes.

```python
pitch = VerticalPitch(
    half=True,
    pad_left=-3,  # bring the left axis in 10 data units (reduce the size)
    pad_right=-3,  # bring the right axis in 10 data units (reduce the size)
    pad_top=0,  # extend the top axis 10 data units
    pad_bottom=0)  # extend the bottom axis 20 data units
fig, ax = pitch.draw()
```

## Pitch appearance

Use `pitch_color` and `line_color`, and `stripe_color` (if `stripe=True`) to adjust the colors.

```python
pitch = Pitch(pitch_color='#aabb97', line_color='white',
              stripe_color='#c2d59d', stripe=True)  # optional stripes
fig, ax = pitch.draw()
```

### Line style

Use `linestyle` and `goal_linestyle` to adjust the colors.

```python
pitch = Pitch(linestyle='--', linewidth=1, goal_linestyle='-')
fig, ax = pitch.draw()
```

### Line alpha

Use `pitch_alpha` and `goal_alpha` to adjust the colors

```python
pitch = Pitch(line_alpha=0.5, goal_alpha=0.8)
fig, ax = pitch.draw()
```

### Corner arcs

```python
pitch = VerticalPitch(corner_arcs=True, half=True, pad_bottom=-30)
fig, ax = pitch.draw()
```

### Juego de Posición

```python
pitch = Pitch(positional=True, shade_middle=True, positional_color='#eadddd', shade_color='#f2f2f2')
fig, ax = pitch.draw()
```

grass color

```python
pitch = Pitch(pitch_color='grass', line_color='white',
              stripe=True)  # optional stripes
fig, ax = pitch.draw()
```

```python
pitch = Pitch(axis=True, label=True, tick=True)
fig, ax = pitch.draw()
```
