# Setup

```python
# import SBoppen from mplsoccer
from mplsoccer import Sbopen
# create a object called parser to open the data.
parser = Sbopen()
```

# Competition Data

Data Structure:
- competition_id --> season_id

e.g. UCL 2020

```python
# opening data using competition method
competitions = parser.competition()
# check data structure
# competitions.info()
```

```python
# Overview
# competitions.head(50)
competitions.sort_values('season_name',ascending=True).head(40)
```

# Match data

explore further matches of a competition.

e.g. UCL16/17 comp_id=16,season_id=2

```python
# opening data using match method
match = parser.match(competition_id=43, season_id=55)
# check data structure
# match.info()
```

```python
# have a overview
match[['home_team_name','away_team_name','home_score','away_score','match_date']]
# only final match available
```

## Event data


- Event: events with additional information
- Tactics: information about player position on the pitch
- Related: information on events that were related to each other - for example ball pass and pressure applied
- Freeze: freezed frames with player position in the moment of shots

```python
# opening data
# only event data now
event = parser.event(18244)[0]
event.info()
```
